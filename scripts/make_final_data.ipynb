Ce Notebook permet de mettre en forme les données nettoyées pour qu'elles soient prêtes à l'utilisation.

```python
# on va récupérer les avis et les pseudos générés pour en faire des données propres et prête à être utilisées pour la suite du projet : 
import pandas as pd

# on va lire le fichier txt qui contient les avis générés :
df = pd.read_csv('./datas/generated_reviews_clean.txt', sep='\t', header=None, names=['date', 'review'])

# on va afficher les premières lignes du dataframe pour vérifier que les données ont été correctement chargées :
df.head()
```

```python
# Maintenant celui qui contient les pseudos générés :
df_pseudos = pd.read_csv('./datas/pseudos.txt', sep='\t', header=None, names=['pseudo'])

df_pseudos.head()
```

```python
# nombre de lignes dans chaque dataframe :
print(f'Nombre d\'avis générés : {len(df)}')
print(f'Nombre de pseudos générés : {len(df_pseudos)}')
```

```python
# on fusionne les deux dans l'ordre pseudo - date - review et on ne prend que les 2000 premières lignes :
df_final = pd.concat([df_pseudos, df], axis=1).head(2001)

# on affiche le dataframe final pour vérifier que les données ont été correctement fusionnées :
df_final
```

```python
# on enregistre le dataframe final dans un fichier pour pouvoir l'utiliser dans la suite du projet :
df_final.to_csv('./datas/final_reviews.txt', sep='\t')

print('Données finales enregistrées dans final_reviews.txt')
```
